```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
price_data = pd.read_csv("pricebehaviour.csv")
print(price_data.head(10))
```

```python
plt.plot(price_data.Time, price_data.Price)
plt.figure(figsize=(10,5))
plt.show()
```

```python
price_data.plot(kind="line",y="Price", figsize=(10,5), style='b-', legend=False, title="Price Behaviour Random Walk Process")
```

```python
price_data.plot(
    kind="line", subplots=True, layout=(3,2),figsize=(15,10), title="Price Behaviour as a random walk process"
)
```

```python
sns.relplot(kind="line", x="Time", y="Price", data=price_data)
```

```python
from pandas.plotting import bootstrap_plot
bootstrap_plot(price_data.Price, fig=plt.figure(figsize=(10,6)))
```

In a random walk model the prices of securities are studied under the assumption that 
changes are random. That is prices can move up and down by a random value given by a normal distribution. What I have done in this project is an approximation of the complex market behaviour, it's close to real processes based on random walk models.
